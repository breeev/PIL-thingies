```python
img=Image.new('RGB',size=(100,)*2)
w,h=img.size
r,g,b=(255,0,0)
def pix(x,y):
          global r,g,b
          img.putpixel((x,y),(r,g,b))
          for i in range(10):r,g,b=update(r,g,b)
          ok=[p for p in[(x,(y-1)%h),((x-1)%w,y),(x,(y+1)%h),((x+1)%w,y)]if(not img.getpixel(p)[0])]
          if ok:
                    possibilities=[]
                    for i in range(len(ok)):
                              x,y=ok[i]
                              oki=[p for p in[(x,(y-1)%h),((x-1)%w,y),(x,(y+1)%h),((x+1)%w,y)]if(not img.getpixel(p)[0])]
                              if len(oki)==3:possibilities.append((x,y))
                    if possibilities:pix(*possibilities[randint(0,len(possibilities)-1)])
pix(w//2,h//2)
display(img.resize(size=(2000,)*2,resample=Image.Resampling.NEAREST))
```
